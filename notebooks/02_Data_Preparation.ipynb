## Objectives
- Reshape historical sales data from wide to long format
- Merge calendar information
- Merge selling prices

```python
# Load the raw source datasets

calendar, sales, prices = load_data()

print(f"Calendar: {calendar.shape}")
print(f"Sales: {sales.shape}")
print(f"Prices: {prices.shape}")
```

The sales dataset stores historical demand in a wide format with one column per day (`d_1` to `d_1913`).
Converting the entire dataset into long format produces over 58 million records, which can exceed available memory.
To improve reliability and scalability, the transformation is processed in chronological chunks.

Each chunk:
1. Converts wide sales data to long format.
2. Merges calendar information.
3. Merges selling prices.
4. Saves the result as a Parquet file.

After all chunks are processed, they are combined into a single master dataset.

```python
# Transform raw data into the analytical master dataset

create_master_dataset(
    sales=sales,
    calendar=calendar,
    prices=prices,
)
```
